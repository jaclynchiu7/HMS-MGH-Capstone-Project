```python
import pandas as pd
import numpy as np

ADI = "US_2020_ADI_Census_Block_Group_v3.2.csv"
ADI_data = pd.read_csv(ADI, low_memory = False)
ADI_df = pd.DataFrame(ADI_data)
```

```python
ADI_df.head()
```

```python
case = "us_county_confirmed_cases.csv"
case_data = pd.read_csv(case, low_memory = False) 
case_df = pd.DataFrame(case_data)

death = "us_county_confirmed_deaths.csv"
death_data = pd.read_csv(death, low_memory = False)
death_df = pd.DataFrame(death_data)
```

```python
result0 = pd.merge(case_df, death_df, how='inner', on='COUNTY',
                   left_index=False, right_index=False, sort=True,
                   suffixes=('_x', '_y'))
```

- Geocode (FIPS Code)—Up to 12-digit string showing the census geography for the current record. The geocode is in the following format:
-- 1–2—State FIPS Code
-- 3–5—County FIPS Code
-- 6–11—Census Tract/BNA
-- 12—Block Group

```python
ADI_df['COUNTY'] = ''
ADI_df['STATE'] = ''
ADI_df['FIPS'] = ADI_df['FIPS'].astype(str)
for i in range(0, len(ADI_df)):
    ADI_df['FIPS'][i] = ADI_df['FIPS'][i].zfill(12)
    # extract county fips
    ADI_df['COUNTY'][i]= ADI_df['FIPS'][i][0:5]
    # extract state 
    ADI_df['STATE'][i]= ADI_df['FIPS'][i][0:2]
```

```python
ADI_df.head()
```

```python
# achieve average RNK for national and state
ADI_df_clean = ADI_df.drop(['OBJECTID','GISJOIN','FIPS'],axis=1, inplace=False)
```

```python
ADI_df_clean.head()
```

```python
# TODO: delete string in 2020 ADI
def is_number(s):
    """判断字符串是否为数字"""
    try:
        float(s)
        return True
    except ValueError:
        return False

for i in range (0,len(ADI_df_clean)):
    if not is_number(ADI_df_clean['ADI_NATRANK'][i]):
        ADI_df_clean = ADI_df_clean.drop(index = i)
```

```python
type(ADI_df_clean["ADI_NATRANK"][0])
```

```python
ADI_df_clean["ADI_NATRANK"] = ADI_df_clean["ADI_NATRANK"].astype(int)
ADI_df_clean["ADI_STATERNK"] = ADI_df_clean["ADI_STATERNK"].astype(int)
```

```python
ADI_df_clean_agg = ADI_df_clean.groupby('COUNTY').agg({"ADI_NATRANK": np.mean, "ADI_STATERNK": np.mean})
```

```python
# ADI_df['COUNTY'] = ADI_df['COUNTY'].astype(int)
result1 = pd.merge(result0, ADI_df_clean_agg, how='left', on='COUNTY')
```
